Observe the evolution of car fuel consumption (mpg) in relation to horsepower.

```python
#Observe the evolution of car fuel consumption (mpg) in relation to horsepower.
plt.figure(figsize=(11,5))
sns.lmplot(data = df, x = 'mpg', y = 'horsepower', hue = 'model_year', col = 'model_year')
```

What do we seem to see?

#form this lmplot we observ that the variables are connected at all the modul years

```python
df_USA = df[(df['origin'] == 'usa') & (df['model_year']>= 70) & (df['model_year'] <= 82)]
df_USA
```

```python
#Observe the evolution of the average weight of American cars from the year 1970 to the year 1982.

plt.figure(figsize= (11,5))
sns.lineplot(x = df_USA['model_year'], y = df_USA['weight'])
plt.title("AVG of models weight over years", c = 'k', fontsize = 15)
```

```python
# it seems like the weight of cars is being tighty over time
```

What is the average acceleration of cars in different countries?

```python
#What is the average acceleration of cars in different countries?
plt.figure(figsize=(11,5))
sns.barplot(data= df, x = df['origin'], y = df['acceleration'], errorbar= None)
plt.title("AVG of acceleration in every country", c = 'k', fontsize = 15)
```

What is it for models from the year 80?

```python
df_80s = df[df['model_year'] == 80]
df_80s
```

```python
plt.figure(figsize=(11,5))
sns.barplot(data= df_80s, x = df['origin'], y = df['acceleration'], errorbar=None)
plt.title("AVG of acceleration by country at 80's")
```
